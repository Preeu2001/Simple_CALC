```python
from tkinter import*

me=Tk()
me.geometry("354x460")
me.title("CALCULATOR")
melabel = Label(me,text="CALCULATOR",bg='dark gray',font=("Times",30,'bold'))
melabel.pack(side=TOP)
me.config(background='Dark gray')
metext=Entry(me,font=("Courier New",12,'bold'),width=25,bd=5,bg='powder blue')
metext.pack()

def ButtonClick(number):
    current = metext.get()
    metext.delete(0,END)
    metext.insert(1,str(current)+str(number))

def ButtonClear():
    metext.delete(0,END)

def ButtonAdd():
    global fnum,math
    math = 1
    first_number = metext.get()
    metext.delete(0,END)
    fnum = int(first_number)

def ButtonSub():
    global fnum,math
    math = 2
    first_number = metext.get()
    metext.delete(0,END)
    fnum = int(first_number)

def ButtonMul():
    global fnum,math
    math = 3
    first_number = metext.get()
    metext.delete(0,END)
    fnum = int(first_number)

def ButtonDiv():
    global fnum,math
    math = 4
    first_number = metext.get()
    metext.delete(0,END)
    fnum = int(first_number)

def ButtonEqual():
    second_number = metext.get()
    metext.delete(0,END)
    snum = int(second_number)

    if math == 1 :
        metext.insert(2,fnum+snum)

    elif math == 2:
        metext.insert(2,fnum-snum)

    elif math == 3:
        metext.insert(2,fnum*snum)

    else:
        try:
            metext.insert(2,fnum/snum) 
        except:
            metext.insert(2,'Error') 

but1=Button(me,padx=14,pady=14,bd=4,bg='white',text="1",font=("Courier New",16,'bold'),command=lambda : ButtonClick(1))
but1.place(x=10,y=100)

but2=Button(me,padx=14,pady=14,bd=4,bg='white',text="2",font=("Courier New",16,'bold'),command=lambda : ButtonClick(2))
but2.place(x=10,y=170)

but3=Button(me,padx=14,pady=14,bd=4,bg='white',text="3",font=("Courier New",16,'bold'),command=lambda : ButtonClick(3))
but3.place(x=10,y=240)

but4=Button(me,padx=14,pady=14,bd=4,bg='white',text="4",font=("Courier New",16,'bold'),command=lambda : ButtonClick(4))
but4.place(x=75,y=100)

but5=Button(me,padx=14,pady=14,bd=4,bg='white',text="5",font=("Courier New",16,'bold'),command=lambda : ButtonClick(5))
but5.place(x=75,y=170)

but6=Button(me,padx=14,pady=14,bd=4,bg='white',text="6",font=("Courier New",16,'bold'),command=lambda : ButtonClick(6))
but6.place(x=75,y=240)

but7=Button(me,padx=14,pady=14,bd=4,bg='white',text="7",font=("Courier New",16,'bold'),command=lambda : ButtonClick(7))
but7.place(x=140,y=100)

but8=Button(me,padx=14,pady=14,bd=4,bg='white',text="8",font=("Courier New",16,'bold'),command=lambda : ButtonClick(8))
but8.place(x=140,y=170)

but9=Button(me,padx=14,pady=14,bd=4,bg='white',text="9",font=("Courier New",16,'bold'),command=lambda : ButtonClick(9))
but9.place(x=140,y=240)

but0=Button(me,padx=14,pady=14,bd=4,bg='white',text="0",font=("Courier New",16,'bold'),command=lambda : ButtonClick(0))
but0.place(x=10,y=310)

butdot=Button(me,padx=47,pady=14,bd=4,bg='white',text=".",font=("Courier New",16,'bold'))
butdot.place(x=75,y=310)

butpl=Button(me,padx=14,pady=14,bd=4,bg='white',text="+",font=("Courier New",16,'bold'),command=ButtonAdd)
butpl.place(x=205,y=100)

butsub=Button(me,padx=14,pady=14,bd=4,bg='white',text="-",font=("Courier New",16,'bold'),command=ButtonSub)
butsub.place(x=205,y=170)

butml=Button(me,padx=14,pady=14,bd=4,bg='white',text="*",font=("Courier New",16,'bold'),command=ButtonMul)
butml.place(x=205,y=240)

butdiv=Button(me,padx=14,pady=14,bd=4,bg='white',text="/",font=("Courier New",16,'bold'),command=ButtonDiv)
butdiv.place(x=205,y=310)

butclear=Button(me,padx=14,pady=119,bd=4,bg='white',text="CE",font=("Courier New",16,'bold'),command=ButtonClear)
butclear.place(x=270,y=100)

butequal=Button(me,padx=151,pady=14,bd=4,bg='white',text="=",font=("Courier New",16,'bold'),command=ButtonEqual)
butequal.place(x=10,y=380)
me.mainloop()
```
